```python
#In arrays it is not allow  strings, only allow numeric data types int,float
A=np.array([34.2,56.7,78.2,90.9,90.8])
print("Maximum is",A.max())  # returns max element in array
print("minimum is",A.min())  # returns min element in array
```

```python
#Max Function with Axis
A = np.array([34,56,78,90]).reshape(2,2)
print(A)
print("________________")
print(A.max(axis=0)) #axis=0 is column max number
print("________________")
print(A.max(axis=1)) #axis=1 is row maximum number
```

```python
#Min Function with axis
A = np.array([34,56,78,90,16,78,32,19,14,23,97,11]).reshape(3,4)
print(A)
print("________________")
print(A.min(axis=0)) #axis=0 is column minimum number
print("________________")
print(A.min(axis=1)) #axis=1 is row minimum number
```

```python
#Joining Arrays
#Vertical stack- it join the arrays vertically(row wise)
A = np.array([34,56,78,90]).reshape(2,2)
B = np.array([7,9,3,12]).reshape(2,2)
print(A)
print("______________")
print(B)
print("______________")
print("After vertically stacking the arrays")
print(np.vstack((A,B)))  # Syntax np.vstack((a,b))
```

```python
#Horizontal stack- it join the arrays horizontally(column wise)
A = np.array([34,56,78,90]).reshape(2,2)
B = np.array([7,9,3,12]).reshape(2,2)
print(A)
print("______________")
print(B)
print("______________")
print("After Horizontally stacking the arrays")
print(np.hstack((A,B)))  # Syntax np.hstack((a,b))
```

```python
#Axis=0 is vstack function
A = np.array([34,56,78,90]).reshape(2,2)
B = np.array([7,9,3,12]).reshape(2,2)
print(A)
print("______________")
print(B)
print("______________")
print(np.stack((B,A),axis=0)) #axis=0 in stack is vstack format
```

```python
#Dimensionality stack (dstack)  colums to row shifted ,no of rows=no of groups
A= np.arange(30).reshape(2,3,5)  #(g,r,c)
print(A)
print("After dstack")
print(np.dstack(A)) #no.of rows -->no of groups g=r
#no.of column in every group --> no of rows in every group
```

```python
#Splitting arrays vsplit is vertically split
A= np.arange(18).reshape(9,2)
print(A)
print("After splitting")
np.vsplit(A,3) #(split into equal halfs) the number is divisible by row number Ex:9rows/3= 9/3=3 splitt
```

```python
#argmax function- it returns the position of maximum element in array it starts from 0
A=np.arange(1,19).reshape(3,6)
print(A)
print(A.argmax())  # index of highest 18 number is 17
```

```python
# argmax in axis=0(column)
A= np.array([17,4,18,56,12,33,89,21,34,55,11,13]).reshape(3,4)
print(A)
print(A.argmax(axis=0)) #it returns max number position in every column(starts from 0)
```

```python
# argmax in axis=1(row)
A= np.array([17,4,18,56,12,33,89,21,34,55,11,13]).reshape(3,4)
print(A)
print(A.argmax(axis=1)) #it returns max number position in every row(starts from 0)
```

```python
#argmin function- it returns the position of minimum element in array it starts from 0
A=np.arange(1,19).reshape(3,6)
print(A)
print(A.argmin()) #index of lowest element 1 is 0
```

```python
# argmin in axis=0(column)
A= np.array([17,4,18,56,12,33,89,21,34,55,11,13]).reshape(3,4)
print(A)
print(A.argmin(axis=0)) #it returns min number position in every column(starts from 0)
```

```python
# argmin in axis=1(row)
A= np.array([17,4,18,56,12,33,89,21,34,55,11,13]).reshape(3,4)
print(A)
print(A.argmin(axis=1)) #it returns min number position in every row(starts from 0)
```

```python
#Joints Functions
# 1)Inner function- #multiply first element in first array to first element in second array then sum
A=np.array([1,2,3]) #multiply second element in first array to second element in second array then sum....
B=np.array([4,5,6])
# 1*4 + 2*5 + 3*6
print(np.inner(A,B))
```

```python
#2)Outer function - it multiplies  first number in first array with all numbers in second array then sum
                   #multiplies second number in first array with all numbers in second array then sum...
A=np.array([1,2,3])
B=np.array([4,5,6])
# 1*(4,5,6) 2*(4,5,6) 3*(4,5,6)
print(np.outer(A,B))
```

```python
#3)Cross function - it is same as (ad-bc) det value format
A=np.array([1,2,3])
B=np.array([4,5,6])
print(np.cross(A,B))
```
