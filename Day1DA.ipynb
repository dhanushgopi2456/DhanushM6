```python
#take the number from the user and satisfy the below conditions
# sal is 0 to 50000 5% reduction,50000 to 70000 10% reduction,
# 70000 to 100000 12% reduction above 1L next level promoted
a=int(input("Enter the Salary:"))
if(a>=0 and a<50000):
  print("tax applied 5%",a-(a*5/100))
elif(a>=50000 and a<70000):
  print("tax applied 10%",a-(a*10/100))
elif(a>=70000 and a<100000):
  print("tax applied 12%",a-(a*12/100))
else:
  print("sal is more next level promoted")

```

```python
#list comprehension
#newlist = [output for]
l1=[23,34,45,38,12,456];
l2=[i*i for i in l1];
print(l2);

```

```python
#list comprehension
#newlist = [output for if con]
l1=[23,34,45,38,12,67,456];
l2=[i*i for i in l1 if i>50];
print(l2);
```

```python
#list comprehension
#newlist = [(st of if)if (cond) else(st of else) for]
l1=[23,95,38,167,456];
l2=[(i*i) if (i>50) else ('X') for i  in l1 ];
print(l2);
```
