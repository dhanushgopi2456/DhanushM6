```python
# They are 5 subjects psychology,nano technology,medicine,onchology,anthropolgy
# TWo students have applied to do masters in any of these five fields with aline plot show in which field they can exel the cut for each subject is 70
# now student1=[67,45,73,71,57]  student2=[43,67,68,53,85]
sub=["Psy","Nano","Med","One","Anthro"]
#student1
#plt.plottype(X,Y,label,color,marker,linewidth,linestyle)
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]
plt.plot(sub,s1,label="student1",color='k',marker='p') #Student1
plt.plot(sub,s2,label="student2",color='b',marker='p') #Student2
plt.plot(sub,cutoff,label="cutoff",color='r') #cutoff
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
#An India vs Australia is on live india has given 298 as target for australia ,
#the runs for every five overs is given as follows 25,51,84,131,160,189,220,250,267,297
#Austalia have completed 25 overs 15,41,94,110,151
#plot overs vs runs for both the countries with india has blue line and australia has yellow line
#with required labels and legend the marker for india is flag and australia is plus
#Take overs from arange functions create runs as arrays linewidith is 2
import numpy as np
from matplotlib  import pyplot as plt
plt.xlabel("Overs")
plt.ylabel("Runs")
A=np.array([25,51,84,131,160,189,220,250,267,297])
B=np.array([15,41,94,110,151])
o=np.arange(5,51,5)
o1=np.arange(5,26,5)
plt.title("IndIndia vs Aus")
plt.plot(o,A,label="India",color='b',marker=">",linewidth=2)
plt.plot(o1,B,label="Australia",color='y',marker='p',linewidth=2)
plt.legend()
plt.grid()




```
